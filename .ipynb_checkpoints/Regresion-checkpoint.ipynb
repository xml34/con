## REGRESION LINEAL



Se ha decidido aplicar este método debido a las caracteristicas del dataSet empleado, este es una colección de datos con 9 columnas las cuales son:
Date: fecha ddmmaaaa
hora: hora hhmmss
Global_active_power: KW-min (kiloVatios)
Global_reactive_power KW-min
Voltage:  V-min
Global_intensity   A-min   (Amperios)
Sub_metering_1: horno/microondas/lavaplatos (Cocina)
Sub_metering_2: lavadora/secadora/devera/1bombillo  (Lavanderia)
Sub_metering_3: celentadorDeAgua/acondicionador  (Baño)
se puede intuir que hay una relacion directa entre el uso de los electrodomésticos y el consumo de energía
en terminos de (W,V,A)
para confirmar esto usaremos el algoritmo de regresión lineal esperando que se refleje el comportamiento esperado

```python
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import numpy as np
regr=linear_model.LinearRegression()

#SE CARGA EL DATASET
d=pd.read_csv("household_power_consumption.csv")
df=pd.DataFrame(d)

#SE ELIGEN LAS VARIABLES GLOBAL_A_P Y LOS ELECTRODOMESTICOS DE CONCINA
x=df['Global_active_power']
y=df['Sub_metering_1']

X=x[:,np.newaxis]

#SE ENTRENA EL MODELO
print(regr.fit(X,y))
print(regr.coef_)
m=regr.coef_[0]
b=regr.intercept_
y_p=m*X+b
print('y={0}*x+{1}'.format(m,b))
print(regr.predict(X)[0:5])
print("msg",r2_score(y,y_p))

#IMPRIMIR GRAFICA
plt.scatter(x,y,color='blue')
plt.plot(x,y_p,color='red')
plt.title("Vatos vs UsoEnergia",fontsize=16)
plt.xlabel("UsoCocina",fontsize=13)
plt.ylabel("VatiosPromedio",fontsize=13)

```

Al observar la gráfica se podría decir que no existe una relación ideal entre estas dos variables.
Esto se debe a que la potencia promedio es afectada por la sub_medicion_1,2,3; es decir 
todos los elementos electronicos que consumen el el momento.
Ademas segun los datos el uso de los electrodomesticos no se hace todo el tiempo, hay momentos del dia en que no se usan por lo tanto son registrados como "0" lo cual reduce la precision del modelo como se ve reflejado en r^2 
para esto incluiremos las tres sub_mediciones y se tomará los valores > 0

```python
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
regr=linear_model.LinearRegression()

d=pd.read_csv("household_power_consumption.csv")
d = d[d['Sub_metering_1'] > 0]
d = d[d['Sub_metering_2'] > 0]
d = d[d['Sub_metering_3'] > 0]
df=pd.DataFrame(d)
df

x=df['Global_active_power']
y=df['Sub_metering_1']+df['Sub_metering_2']+df['Sub_metering_3']

X=x[:,np.newaxis]
#print(X)
print(regr.fit(X,y))
print(regr.coef_)
m=regr.coef_[0]
b=regr.intercept_
y_p=m*X+b
print('y={0}*x+{1}'.format(m,b))
print(regr.predict(X)[0:5])
print("msg",r2_score(y,y_p))

plt.scatter(x,y,color='blue')
plt.plot(x,y_p,color='red')
plt.title("Vatos vs UsoEnergia",fontsize=16)
plt.xlabel("UsoEnegiaCasa",fontsize=13)
plt.ylabel("VatiosPromedio",fontsize=13)
```

De este modo se puede observar la realacion entre el consumo de energía y el gasto registrado (potencia registrada)
se evidencia que entre mas vatios generados por los elementos electrónicos mayor es el registro de voltaje, Amperaje, y vatios (promedio)

```python
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
regr=linear_model.LinearRegression()

d=pd.read_csv("household_power_consumption.csv")
d = d[d['Sub_metering_1'] > 0]
d = d[d['Sub_metering_2'] > 0]
d = d[d['Sub_metering_3'] > 0]
df=pd.DataFrame(d)
df

x=df['Global_intensity']
y=df['Sub_metering_1']+df['Sub_metering_2']+df['Sub_metering_3']

X=x[:,np.newaxis]
#print(X)
print(regr.fit(X,y))
print(regr.coef_)
m=regr.coef_[0]
b=regr.intercept_
y_p=m*X+b
print('y={0}*x+{1}'.format(m,b))
print(regr.predict(X)[0:5])
print("msg",r2_score(y,y_p))

plt.scatter(x,y,color='blue')
plt.plot(x,y_p,color='red')
plt.title("Amperaje vs UsoEnergia",fontsize=16)
plt.xlabel("UsoEnergiaCada",fontsize=13)
plt.ylabel("Amperaje",fontsize=13)
```
