## Clusterización



Cotinuando con el analisis, tambien se puede intuir que debe existir una realción entre la hora del dia y el consumo de energía, intuitivamente esta relación puede revelar las horas del día en que mas se utiliza la energía por seccion del hogar (cocina,lavadero,baño)
para esto se cree que es conveniente utlizar la clusterisacion o agrupación ya que se esperan que los datos se acumules (segmenten) en horas del día

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

datos=pd.read_csv("prueba.csv")
df=pd.DataFrame(datos)

x=df['Time'].values
y=df['Sub_metering_1'].values

#info=df[['Date','Sub_metering_1']].as_matrix()
#print(info)

X=np.array(list(zip(x,y)))
#print(X)

kmeans=KMeans(n_clusters=5)
kmeans=kmeans.fit(X)
labels=kmeans.predict(X)
centroids=kmeans.cluster_centers_

colors=["m.","r.","c.","y.","b."]

for i in range(len(X)):
    #print("Coodenada: ",X[i]," Label: ",labels[i])
    plt.plot(X[i][0],X[i][1],colors[labels[i]],markersize=10)
plt.scatter(centroids[:,0],centroids[:,1],marker='x',s=150,linewidths=5,zorder=10)
plt.show()
    
```

Para la primera variable se observa que los horas del dia en que se hace mayor uso de la energía en la cocina son 3:
1. 8:30-10:00
2. 10:00-12
3. 15:00-17:00

donde 1 y 2 podrian ser parte de un mismo cluster
